```python
age = int(input("Enter your age: "))
print("You entered:", age)
future_age = age + 5
print(f"After five years you will be = {age} + 5 = {future_age}")
```

```python
## Simple Interest Calculator
```

```python
p = float(input("Principal: "))
r = float(input("Rate (annual %): "))
t = float(input("Time (years): "))
si = (p * r * t) / 100
print(f"Formula: Simple Interest = (p * r * t) / 100")
print(f"Computation: ({p} * {r} * {t}) / 100 = {si}")
print("Simple interest =", si)
```

```python
## Ohm's law calculator
```

```python
resistance = float(input("Resistance (ohms): "))
voltage = float(input("Voltage (volts): "))
current = voltage / resistance if resistance != 0 else None
print(f"Formula: Current = Voltage / Resistance")
if current is None:
    print("Resistance is zero — current is infinite or undefined")
else:
    print(f"Computation: {voltage} / {resistance} = {current}")
    print("Current =", current)
```
