```python
grocery = ["Rice = 70","Milk = 80","Eggs = 90","Cabbage = 100"]
print("Grocery List")
number = 1
for item in grocery:
  print(f"{number}.{item}")
  number = number+1

rice_price = 70
milk_price = 80
egg_price = 90
cabbage_price = 100

def calculate_total(price1,price2,price3,price4):
    return price1+price2+price3+price4
print(f"Total = {calculate_total(rice_price,milk_price,egg_price,cabbage_price)}")
```
