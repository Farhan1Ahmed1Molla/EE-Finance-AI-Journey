```python
##Football ticket receipt
```

```python
Quantity = int(input("Number of tickets = "))

Ticket_price = 50

def calculate_total():
   return Ticket_price * Quantity

calculate_total()
print(f"Total = {calculate_total()}")

total = calculate_total()

if total > 200:
   print("BULK BOOKING")

else:
   print("REGULAR BOOKING")
```

```python
## Football squad manager
```

```python
Players = ["Tommy","Halaand","Ronaldo","Yamael"]

number = 1

for item in Players :
   print(f"{number}.{item}")
   number += 1

player = input("Enter player name =  ")

def check_player(item):
   return item in Players

if check_player(player):
   print("Player found")

else:
   print("Player not found")

Rating = int(input("Enter Rating: "))


if Rating >= 90:
   print("World Class")

elif Rating >= 80:
   print("Excellent")

elif Rating >= 70:
   print("Good")

elif Rating >= 60:
   print("Average")

else:
   print("Not qualified")
```
