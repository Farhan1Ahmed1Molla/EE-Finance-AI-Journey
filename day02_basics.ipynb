```python
## Adult marker
```

```python
age = int(input("Enter your age:"))
if age >= 18:
 print("you are an adult")
elif age >= 12:
 print("you are a teenager")
elif age >= 4:
 print(" you are a child")
else:
 print("you are a baby")
```

```python
## pass/fail and grading system
```

```python
marks = int(input("Enter your marks:"))
if marks >= 80:
 print(" you pass= A",marks)
elif marks >= 60:
 print(" you pass= B",marks)
elif marks >= 40:
 print(" you pass= C",marks)
else:
 print(" you fail= D",marks)
```
