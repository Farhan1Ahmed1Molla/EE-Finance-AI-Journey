```python
## Ohm's law calculator
```

```python
resistance = float(input("Resistance (ohms): "))
voltage = float(input("Voltage (volts): "))
current = voltage / resistance if resistance != 0 else None
print(f"Formula: Current = Voltage / Resistance")
if current is None:
    print("Resistance is zero — current is infinite or undefined")
else:
    print(f"Computation: {voltage} / {resistance} = {current}")
    print("Current =", current)
```
